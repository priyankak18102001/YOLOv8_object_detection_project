## Detection + Counting Loop

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load trained model
model = YOLO("runs/detect/train2/weights/best.pt")

video_path = r".\TestVideo\TestVideo.mp4\TestVideo\TrafficPolice.mp4"
cap = cv2.VideoCapture(video_path)

# Output video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("vehicle_count_output.mp4", fourcc, 20.0,
                      (int(cap.get(3)), int(cap.get(4))))

# Dictionary for total counts across video
total_counts = {}

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    frame_counts = {}

    for r in results:
        boxes = r.boxes
        names = r.names
        annotated_frame = r.plot()

        for box in boxes:
            class_id = int(box.cls[0])
            class_name = names[class_id]

            # Count per frame
            if class_name not in frame_counts:
                frame_counts[class_name] = 0
            frame_counts[class_name] += 1

            # Count total across video
            if class_name not in total_counts:
                total_counts[class_name] = 0
            total_counts[class_name] += 1

    # Display class-wise counts on frame
    y_pos = 30
    for key, value in frame_counts.items():
        text = f"{key}: {value}"
        cv2.putText(annotated_frame, text, (20, y_pos),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
        y_pos += 30

    out.write(annotated_frame)

cap.release()
out.release()

print("Video saved successfully!")
print("Total Vehicle Counts:", total_counts)
```

### Create Graph of Vehicle Counts

```python
%matplotlib inline

classes = list(total_counts.keys())
values = list(total_counts.values())

plt.figure()
plt.bar(classes, values)
plt.title("Total Vehicle Count by Class")
plt.xlabel("Vehicle Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
